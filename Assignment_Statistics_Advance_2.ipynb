1. What is hypothesis testing in statistics?

Ans. Hypothesis testing is a statistical method used to decide whether there's enough evidence in a sample to support a specific claim about a population.

Null Hypothesis (H‚ÇÄ): No effect or difference (status quo).

Alternative Hypothesis (H‚ÇÅ): What you're testing for.

Significance Level (Œ±): The threshold for rejecting H‚ÇÄ (commonly 0.05).

P-value: Probability of observing the result if H‚ÇÄ is true.

If p ‚â§ Œ±, reject H‚ÇÄ.

If p > Œ±, do not reject H‚ÇÄ.

It helps make data-driven decisions or test claims.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans. Null hypothesis (H‚ÇÄ): Assumes no effect or no difference.

Alternative hypothesis (H‚ÇÅ or Ha): Assumes there is an effect or difference.

Difference: H‚ÇÄ is the default; H‚ÇÅ is what you're trying to prove.

3. What is the significance level in hypothesis testing, and why is it important?

Ans. Significance level (Œ±): The probability of rejecting the null hypothesis when it is true (Type I error).

Common value: 0.05

Importance: Sets the threshold for deciding if results are statistically significant.

4. What does a P-value represent in hypothesis testing?

Ans. P-value: The probability of getting the observed results (or more extreme) if the null hypothesis is true.

Low p-value (‚â§ Œ±): Strong evidence against H‚ÇÄ ‚Üí reject H‚ÇÄ.

High p-value (> Œ±): Weak evidence ‚Üí fail to reject H‚ÇÄ.

5. How do you interpret the P-value in hypothesis testing?

Ans. P ‚â§ Œ±: Reject the null hypothesis (evidence supports H‚ÇÅ).

P > Œ±: Fail to reject the null hypothesis (not enough evidence).
It tells how likely the result is if H‚ÇÄ is true.

6. What are Type 1 and Type 2 errors in hypothesis testing?

Ans. Type 1 Error (Œ±): Rejecting H‚ÇÄ when it is true (false positive).

Type 2 Error (Œ≤): Failing to reject H‚ÇÄ when H‚ÇÅ is true (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans. One-tailed test: Tests for effect in one direction (e.g., greater than).

Two-tailed test: Tests for effect in both directions (e.g., not equal to).
Use depends on the hypothesis.

8. What is the Z-test, and when is it used in hypothesis testing?

Ans. Z-test: A statistical test used when the population variance is known and the sample size is large (n ‚â• 30).

Used for: Testing means or proportions with normal distribution.

10.  What is the T-distribution, and when should it be used instead of the normal distribution?

Ans. T-distribution: A probability distribution similar to normal but with heavier tails.

Use it when:

Sample size is small (n < 30)

Population standard deviation is unknown.

11. What is the difference between a Z-test and a T-test?

Ans. Z-test: Uses known population standard deviation, large samples (n ‚â• 30).

T-test: Uses sample standard deviation, small samples (n < 30), or unknown population std.

12. What is the T-test, and how is it used in hypothesis testing?

Ans. T-test: A test that compares sample mean to a population mean when the population standard deviation is unknown.

Used for: Small samples to check if the mean differs from a value or between groups.

13. What is the relationship between Z-test and T-test in hypothesis testing?

Ans. The T-test approaches the Z-test as sample size grows large because the sample standard deviation better estimates the population standard deviation.

Use T-test for small samples/unknown std dev.

Use Z-test for large samples/known std dev.

14.  What is a confidence interval, and how is it used to interpret statistical results?

Ans. Confidence interval: A range of values likely containing the true population parameter with a certain confidence level (e.g., 95%).

Use: To estimate parameters and assess precision; if it excludes a value (like zero), results may be significant.

15. What is the margin of error, and how does it affect the confidence interval?

Ans. Margin of error: The maximum expected difference between the sample estimate and the true population value.

Effect: Larger margin ‚Üí wider confidence interval; smaller margin ‚Üí narrower interval, meaning more precise estimates.

16.  How is Bayes' Theorem used in statistics, and what is its significance?

Ans. Bayes' Theorem: Updates the probability of a hypothesis based on new evidence.

Significance: Combines prior knowledge with data to improve decision-making and predictions.

17. What is the Chi-square distribution, and when is it used?

Ans. Chi-square distribution: A probability distribution used for testing relationships between categorical variables.

Used in: Chi-square tests for goodness-of-fit and independence in contingency tables.

Applied by: Comparing observed counts to expected counts, then calculating a Chi-square statistic to see if differences are due to chance.

19. What is the F-distribution, and when is it used in hypothesis testing?

Ans. F-distribution: A probability distribution used to compare variances between groups.

Used in: ANOVA and tests comparing two variances.

20. What is an ANOVA test, and what are its assumptions?

Assumptions:

Independent samples

Normally distributed groups

Equal variances (homogeneity of variance)

Two-way ANOVA: Tests two factors and their interaction.

Repeated Measures ANOVA: Tests the same subjects under different conditions/time points.

3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Outcome (0 or 1)
    p (float): Probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = [12, 15, 14, 10, 13, 12, 11, 14, 13, 12]
sample_mean = np.mean(sample)
sample_size = len(sample)

# Known population parameters (for Z-test)
population_mean = 10
population_std = 2

# Significance level
alpha = 0.05

# Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

16.  Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

# Sample data
data = [12, 15, 14, 10, 13, 12, 11, 14, 13, 12]

# Parameters
confidence_level = 0.95
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # standard error of the mean

# Calculate margin of error
h = std_err * stats.t.ppf((1 + confidence_level) / 2, n - 1)

# Confidence interval
lower_bound = mean - h
upper_bound = mean + h

print(f"Sample Mean = {mean:.2f}")
print(f"{confidence_level*100:.0f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z-statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z_stat:.2f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: sample mean is significantly different from population mean.")
    else:
        print("Fail to reject the null hypothesis: no significant difference between sample and population mean.")

# Example usage
sample_data = [12, 15, 14, 10, 13, 12, 11, 14, 13, 12]
population_mean = 10
population_std = 2  # known population std dev

z_test(sample_data, population_mean, population_std)
```

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed p-value

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: sample mean is significantly different.")
    else:
        print("Fail to reject the null hypothesis: no significant difference.")

# Example usage
sample_data = [11, 13, 12, 10, 14, 12, 11, 13, 12, 12]
population_mean = 10
population_std = 2

one_sample_z_test(sample_data, population_mean, population_std)
```

4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z_stat:.2f}")
    print(f"P-value = {p_value:.4f}")

    # Critical z-values for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)
    print(f"Critical z-values: ¬±{z_crit:.2f}")

    # Decision
    if abs(z_stat) > z_crit:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.plot(x, y, label='Standard Normal PDF')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.3)

    # Mark z-statistic
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')

    plt.title('Two-tailed Z-test Decision Regions')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example data
sample_data = [11, 13, 12, 10, 14, 12, 11, 13, 12, 12]
population_mean = 10
population_std = 2

two_tailed_z_test(sample_data, population_mean, population_std)
```

5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(pop_mean, pop_std, sample_size, alpha, true_mean):
    # Critical z-value for alpha (two-tailed)
    z_crit = norm.ppf(1 - alpha/2)

    # Standard error
    se = pop_std / np.sqrt(sample_size)

    # Critical values in sample mean scale
    lower_crit = pop_mean - z_crit * se
    upper_crit = pop_mean + z_crit * se

    # Plot x range covering both distributions
    x = np.linspace(pop_mean - 4*se, true_mean + 4*se, 1000)

    # Null distribution PDF
    null_pdf = norm.pdf(x, pop_mean, se)

    # Alternative distribution PDF (true mean)
    alt_pdf = norm.pdf(x, true_mean, se)

    plt.plot(x, null_pdf, label='Null Hypothesis', color='blue')
    plt.plot(x, alt_pdf, label='Alternative Hypothesis', color='green')

    # Shade Type 1 error region (alpha) under null
    x_alpha_right = np.linspace(upper_crit, pop_mean + 4*se, 100)
    x_alpha_left = np.linspace(pop_mean - 4*se, lower_crit, 100)
    plt.fill_between(x_alpha_right, 0, norm.pdf(x_alpha_right, pop_mean, se), color='red', alpha=0.3, label='Type 1 Error (Œ±)')
    plt.fill_between(x_alpha_left, 0, norm.pdf(x_alpha_left, pop_mean, se), color='red', alpha=0.3)

    # Shade Type 2 error region (beta) under alternative
    beta_region = np.linspace(lower_crit, upper_crit, 100)
    plt.fill_between(beta_region, 0, norm.pdf(beta_region, true_mean, se), color='orange', alpha=0.3, label='Type 2 Error (Œ≤)')

    # Plot critical values
    plt.axvline(lower_crit, color='black', linestyle='--', label='Critical Values')
    plt.axvline(upper_crit, color='black', linestyle='--')

    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_type1_type2_errors(
    pop_mean=50,        # Null hypothesis mean
    pop_std=10,         # Population std dev
    sample_size=30,
    alpha=0.05,         # Significance level
    true_mean=55        # Actual mean under alternative hypothesis
)
```

6. Write a Python program to perform an independent T-test and interpret the results.

8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import norm, t

# Simulate data
np.random.seed(0)
sample_size = 30
pop_mean = 100
pop_std = 15  # population std dev (known)

# Generate sample data from normal distribution
sample = np.random.normal(loc=102, scale=pop_std, size=sample_size)

# Sample stats
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample std dev

# Z-test (population std dev known)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
z_p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# T-test (population std dev unknown)
t_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(sample_size))
t_df = sample_size - 1
t_p_value = 2 * (1 - t.cdf(abs(t_stat), df=t_df))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-test: z = {z_stat:.3f}, p-value = {z_p_value:.4f}")
print(f"T-test: t = {t_stat:.3f}, p-value = {t_p_value:.4f}")

# Interpretation
alpha = 0.05
def interpret_test(p_val):
    return "Reject H0 (significant)" if p_val < alpha else "Fail to reject H0 (not significant)"

print("Z-test result:", interpret_test(z_p_value))
print("T-test result:", interpret_test(t_p_value))
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage
sample_data = [12, 15, 14, 10, 13, 12, 11, 14, 13, 12]
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # t critical value
    return t_crit * std_err

# Example usage
sample_data = [12, 15, 14, 10, 13, 12, 11, 14, 13, 12]
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe:.3f}")
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    prior: P(Disease)
    likelihood: P(Pos | Disease)
    false_positive_rate: P(Pos | No Disease) = 1 - specificity
    """
    # P(No Disease)
    prior_neg = 1 - prior

    # Total probability of positive test: P(Pos)
    p_pos = likelihood * prior + false_positive_rate * prior_neg

    # Posterior: P(Disease | Pos)
    posterior = (likelihood * prior) / p_pos
    return posterior

# Given values
P_disease = 0.01
P_pos_given_disease = 0.95
P_pos_given_no_disease = 1 - 0.90  # 0.10 false positive rate

posterior_prob = bayes_theorem(P_disease, P_pos_given_disease, P_pos_given_no_disease)
print(f"Probability of disease given positive test: {posterior_prob:.4f}")
```

12.  Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table: rows = variable A categories, columns = variable B categories
# Suppose data counts of observations in each category combination
contingency_table = np.array([
    [30, 10, 20],
    [20, 20, 15],
    [10, 15, 25]
])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2:.3f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Reject null hypothesis: variables are dependent (association exists).")
else:
    print("Fail to reject null hypothesis: variables are independent.")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

```python
import numpy as np

def expected_frequencies(observed):
    # Row sums and column sums
    row_sums = observed.sum(axis=1).reshape(-1, 1)
    col_sums = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    # Expected frequency = (row_sum * col_sum) / total
    expected = (row_sums @ col_sums) / total
    return expected

# Example observed data
observed_data = np.array([
    [30, 10, 20],
    [20, 20, 15],
    [10, 15, 25]
])

expected = expected_frequencies(observed_data)
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected)
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (e.g., counts in categories)
observed = np.array([50, 30, 20])

# Expected frequencies (must sum to same total as observed)
expected = np.array([40, 40, 20])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: observed distribution differs from expected.")
else:
    print("Fail to reject null hypothesis: observed distribution matches expected.")
```
